```python
import pandas as pd

# 1. Load your CSV file
df = pd.read_csv('Untitled spreadsheet - house_prices.csv')

# 2. Create equal-sized categories (THIS replaces your function)
df['Price_Category'] = pd.qcut(
    df['price'],
    q=3,
    labels=["Mid_Range", "Premium", "Luxury"]
)

# 3. Save to new file
df.to_csv('house_prices.csv', index=False)

print("File saved successfully as house_prices.csv!")
```

```python
import matplotlib.pyplot as plt

# Count each category
counts = df['Price_Category'].value_counts()

# Step 2: Plot bar chart
counts.plot(kind='bar')

plt.title("Distribution of Price Categories")
plt.xlabel("Category")
plt.ylabel("Number of Houses")
plt.xticks(rotation=0)

plt.show()
```
